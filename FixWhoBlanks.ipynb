# Fix Who Blanks in Emerging Variants Report

```python
import pandas as pd
```

```python
variants = pd.read_csv("July_2023_World_Variants.txt", sep = '\t')
```

```python
variants.head()
```

```python
variants['WHO Label'] = variants['WHO Label'].fillna('Unassigned')
variants['Classification'] = variants['Classification'].fillna('Unassigned')
```

```python
variants.to_csv("July_2023_World_Variants.txt", sep = '\t', index = False)
```
